```python
import pandas as pd
# import user data
u_cols =  ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('../Dataset/ml-100k/u.user', sep='|', names=u_cols, encoding='latin-1')
n_users = users.shape[0]
print('Number of users:', n_users)
users.head(10)
```

```python
# import genre data
g_col = ['genre_name','genre_id']
genre = pd.read_csv('../Dataset/ml-100k/u.genre', sep='|', names=g_col, encoding='latin-1')
n_genre = genre.shape[0]
print('Number of genre:', n_genre)
genre.head(10)
```
