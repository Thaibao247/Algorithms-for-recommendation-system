Create item profiles (feature vector)


```python
X0 = movies.values
X_train_counts = X0[:, -19:]
print(X_train_counts)
```

Use TF-IDF in sklearn library


```python
from sklearn.feature_extraction.text import TfidfTransformer
transformer = TfidfTransformer(smooth_idf=True, norm ='l2')
tfidf = transformer.fit_transform(X_train_counts.tolist()).toarray()
```

```python
for x in range(5):
    print("feature vector for movie ",x)
    print(tfidf[x])
```

```python
# Create function get items that rated by per user
import numpy as np
def get_items_rated_by_user(rate_matrix, user_id):
    y = rate_matrix[:,0] 
    ids = np.where(y == user_id +1)[0] 
    item_ids = rate_matrix[ids, 1] - 1 
    scores = rate_matrix[ids, 2]
    return (item_ids, scores)
```

```python
rate_train = train_data.values
```

```python
rate_test = test_data.values
```

Học mô hình cho mỗi user với L2 Regularization


```python
from sklearn.linear_model import Ridge
from sklearn import linear_model
start_time_cb = time.time()
d = tfidf.shape[1] 
W = np.zeros((d, n_users))
b = np.zeros((1, n_users))
for n in range(n_users):    
    ids, scores = get_items_rated_by_user(rate_train, n)
    clf = Ridge(alpha= 0.01, fit_intercept  = True)
    Xhat = tfidf[ids, :]
    
    clf.fit(Xhat, scores)  
    W[:, n] = clf.coef_  # ndarray of shape (n_features,) or (n_targets, n_features)
    b[0, n] = clf.intercept_ # float or ndarray of shape (n_targets,)
# Use W and b to finish utility matrix
Yhat = tfidf.dot(W) + b
end_time_cb = time.time()-start_time_cb
print("RMSE test: ",RMSE(Yhat,rate_test))
print('time need',end_time_cb)
```

Create training set and test set

```python
ratings_dict = {'item_id': list(ratings.movie_id),
                'user_id': list(ratings.user_id),
                'rating': list(ratings.rating)}
```

```python
df = pd.DataFrame(ratings_dict)
```

```python
reader = Reader(rating_scale=(0.5, 5.0))

data = Dataset.load_from_df(df[['user_id', 'item_id', 'rating']], reader)
```

```python
import time
```

Create model

```python
start_time = time.time()
trainset, testset = train_test_split(data, test_size=0.25)
algo = SVD(n_factors=35, n_epochs=25, lr_all=0.008, reg_all=0.08)
predictions = algo.fit(trainset).test(testset)
rmse_SVD = rmse(predictions)
end_time_svd = time.time()- start_time
print('Time need:', end_time_svd)
print('RMSE SVD:', rmse_SVD)
```

#### Model Based using ALS

Spark Session

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('recommendation').getOrCreate()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
```

```python
data = spark.read.csv('./ratings.csv', inferSchema=True, header=True)
```

```python
## đếm các giá trị null
from pyspark.sql.functions import col,sum
data.select(*(sum(col(c).isNull().cast("int")).alias(c) for c in data.columns)).show()
```

```python
## chia tập dữ liệu thành train và test với 80% cho train và 20% cho test 1 cách ngẫu nhiên
train_data, test_data = data.randomSplit([0.8, 0.2])
```

```python
## xây dựng mô hình gợi ý trên tập dữ liệu train
start_time_als = time.time()
als = ALS(maxIter=10, regParam=0.1, rank=5, nonnegative=True, coldStartStrategy="drop",\
          userCol='user_id', itemCol='movie_id', ratingCol='rating')
model = als.fit(train_data)
predictions = model.transform(test_data)
end_time_als = time.time() - start_time_als
# Kiểm tra root squared error
evaluator = RegressionEvaluator(metricName='rmse', predictionCol='prediction', labelCol='rating')
rmse_als = evaluator.evaluate(predictions)
print('Root mean squared error of the test_data: %.4f' % rmse_als)
print('time need',end_time_als)
```
